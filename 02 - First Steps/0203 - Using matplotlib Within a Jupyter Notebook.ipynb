Then, let's plot a simple line.

```python
plt.plot(range(10))
```

The `plot` command worked, but nothing was displayed, and there's good reason for that. Matplotlib has two working modes: interactive on and interactive off. By default, the interactive mode is off and so we'll need to `plt.show()` in order to display any of the plots that we've made. So, to show the current figure, I now just need to call `plt.show()` like so.

```python
plt.show()
```

Notice that the figure still didn't show up in this notebook. Instead it was displayed in a separate window. The reason is matplotlib is using a different background than one that is supported in a Jupyter notebook. To see which backend we're currently using, just call the `plt.get_backend` function.

```python
plt.get_backend()
```
